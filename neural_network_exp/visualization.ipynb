```python
import numpy as np
import matplotlib.pyplot as plt
import data

mat = data.read_and_create_user_LibCount()

plt.figure(1, figsize=(15, 6))
plt.xlabel('Users')
plt.ylabel('Papers in library')
plt.title('Number of papers in users library')
plt.plot(mat)

plt.figure(2, figsize=(15, 6))
plt.xticks(rotation=55)
binwidth = 5
bins=np.arange(min(mat), max(mat) + binwidth, binwidth)
plt.hist(mat,bins, color='c', edgecolor='black', linewidth=0.2)
plt.xlabel('Papers in library')
plt.ylabel('Users')
plt.grid(True)
plt.yticks(np.arange(0, 1600, step=100))
plt.title('UserLibrary Size Distribution Histogram'.format(binwidth))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import data
mat = data.read_and_create_paper_UserLibFreqCount()

plt.figure(1, figsize=(15, 6))
plt.xlabel('Papers')
plt.ylabel('#Users')
plt.title('Number of times paper appears in users library')
plt.plot(mat)

plt.figure(2, figsize=(15, 6))
binwidth = 5
bins=np.arange(min(mat), max(mat) + binwidth, binwidth)
plt.hist(mat,bins, color='c', edgecolor='black', linewidth=0.2)
plt.title('Paper Frequency Distribution Histogram'.format(binwidth))
plt.grid(True)
plt.yticks(np.arange(0, 8500, step=500))
plt.xlabel('Frequency')
plt.ylabel('Papers')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import data

mat = data.read_and_create_paper_wordFreqCount()
plt.figure(1, figsize=(15, 6))
plt.xlabel('Papers')
plt.ylabel('Vocab Size')
plt.title('Vocab Size of papers in library')
plt.plot(mat)

plt.figure(2, figsize=(15, 6))
plt.yticks(np.arange(0, 1000, step=50))
plt.xticks(np.arange(0, 1500, step=50))
plt.hist(mat, bins = 'auto', color='c', edgecolor='black', linewidth=0.2)
plt.xlabel('Vocab Size')
plt.ylabel('Papers')
plt.grid(True)
plt.title('Paper by Vocab Size Distribution Histogram')
plt.show()
```

```python
import data
import matplotlib.pylab as plt
import matplotlib.patches as mpatches
rd1 = data.get_cruve_readings('loss_plot__2018-04-09T02:48:29.095769.dat')
rd2 = data.get_cruve_readings('loss_plot__2018-04-10T01:13:29.623213.dat')
plt.figure(1, figsize=(15, 6))
plt.plot(rd1, 'C1')
plt.plot(rd2, 'C2')

c1 = mpatches.Patch(color='C1', label='Batch 60')
c2 = mpatches.Patch(color='C2', label='Batch 40')
plt.legend(handles=[c1, c2])

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.yticks(np.arange(0, 7, step=0.5))
plt.grid(True)
plt.show()
```
